```python
sizemlf = 2
def crosssec_plotter(data, ds, idx, xmin, colormap, cmin, cmax, axis, titlestr):
    # zmthetao - dsrecthetao.isel(i=1)
  
    im = data.isel(lev=range(0, 75)).isel(j=range(xmin, 75)).plot(cmap = colormap, vmin = cmin, vmax = cmax, ax=axis) #
    
    # y-axis
    axis.invert_yaxis()
    axis.tick_params(axis='y', labelsize=12)
    axis.set_ylabel('Ocean depth [m]', fontsize=14)
    # x-axis
    axis.set_xticks(range(xmin, 75, 10))
    axis.set_xticklabels(np.round(ds.latitude.isel(i=idx, j=range(xmin,75,10)).values,1), fontsize=12)
    axis.set_xlabel('Longitude [°N]', fontsize=14)
    # title
    axis.set_title(titlestr)
    axis.title.set_size(16) 
    # colorbar
    # cbar = fig.colorbar(im)
    # axis.tick_params(axis='both', labelsize = 14)

    # cbar.tick_params(labelsize=20)
```
